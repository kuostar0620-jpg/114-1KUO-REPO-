```python
# å®‰è£å¥—ä»¶
!pip install gradio pandas -q

import pandas as pd
import gradio as gr

# === Google Sheet é€£çµ ===
sheet_url = "https://docs.google.com/spreadsheets/d/1PrREt6zIAhDwzDdaMmFibUYr_I6d8z24rkdkei_QyYs/edit?gid=1287423805"
csv_url = sheet_url.replace("/edit?gid=", "/export?format=csv&gid=")

# === è®€å–è³‡æ–™ ===
df = pd.read_csv(csv_url)

# === è¡Œå‰æé†’æ¨¡æ¿è¨­å®šï¼ˆå¯è‡ªè¡Œæ”¹ï¼‰===
reminders = {
    "æœƒè¨ˆ": "è«‹å¸¶è¨ˆç®—æ©Ÿï¼Œä¸¦é ç¿’ä¸Šæ¬¡ç·´ç¿’é¡Œã€‚",
    "çµ±è¨ˆ": "è«‹å¸¶ç­†é›»èˆ‡çµ±è¨ˆè¡¨ï¼Œé ç¿’ç¬¬4ç« è¿´æ­¸åˆ†æã€‚",
    "ç¶²éš›ç¶²è·¯ç¨‹å¼è¨­è¨ˆ": "è«‹å¸¶ç­†é›»èˆ‡USBï¼Œè¤‡ç¿’HTMLèˆ‡CSSã€‚",
    "ä¼æ¥­ç®¡ç†": "è«‹é–±è®€æ¡ˆä¾‹åˆ†æç¬¬ä¸‰ç« ã€‚",
    "ç®¡ç†å­¸": "è«‹æº–å‚™å°çµ„å ±å‘Šè³‡æ–™ã€‚",
    "ç¨‹å¼èªè¨€": "è«‹å¸¶ç­†é›»ï¼Œé ç¿’if/elseèˆ‡loopã€‚",
    "ç¶“æ¿Ÿå­¸": "è«‹å¸¶ç­†è¨˜æœ¬ï¼Œé ç¿’ä¾›éœ€æ›²ç·šã€‚",
    "è‹±æ–‡": "è«‹å¸¶èª²æœ¬èˆ‡ç­†è¨˜æœ¬ï¼Œé ç¿’ç¬¬6èª²æ–‡ç« ã€‚",
    "ä½œæ¥­ç³»çµ±": "è«‹å¸¶ç­†é›»ä¸¦é ç¿’process schedulingç« ç¯€ã€‚"
}

# === ç”¢ç”Ÿè¡Œå‰æé†’çš„å‡½å¼ ===
def make_reminder(course, classroom):
    for key, tip in reminders.items():
        if key in course:
            return f"ä»Šå¤©æœ‰ã€Œ{course}ã€èª²ï¼Œåœ°é» {classroom}ï¼Œ{tip}"
    return f"ä»Šå¤©æœ‰ã€Œ{course}ã€èª²ï¼Œåœ°é» {classroom}ï¼Œè«‹æº–æ™‚åˆ°èª²ï¼"

# === æŸ¥è©¢å‡½å¼ ===
def show_schedule(weekday):
    filtered = df[df["æ˜ŸæœŸ"] == weekday]
    if filtered.empty:
        return f"æ˜ŸæœŸ{weekday} æ²’æœ‰èª²ç¨‹ã€‚"

    result = f"ğŸ“… æ˜ŸæœŸ{weekday}çš„èª²ç¨‹ï¼š\n\n"
    for _, row in filtered.iterrows():
        course = row['èª²ç¨‹åç¨±']
        classroom = row['æ•™å®¤']
        result += f"{row['ä¸Šèª²æ™‚é–“']}ï½œ{course}ï¼ˆ{classroom}ï¼‰\n"
        result += f"ğŸ‘‰ è¡Œå‰æé†’ï¼š{make_reminder(course, classroom)}\n\n"
    return result

# === Gradio ä»‹é¢ ===
with gr.Blocks() as demo:
    gr.Markdown("## ğŸ—“ï¸ æŸ¥è©¢æ¯æ—¥èª²è¡¨ + è¡Œå‰æé†’")
    weekday_input = gr.Dropdown(["ä¸€", "äºŒ", "ä¸‰", "å››", "äº”"], label="é¸æ“‡æ˜ŸæœŸ", value="ä¸€")
    output_text = gr.Textbox(label="æŸ¥è©¢çµæœ", lines=20)
    weekday_input.change(show_schedule, inputs=weekday_input, outputs=output_text)

demo.launch()
```
